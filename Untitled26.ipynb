```python
# ======================
# 1. Ø§Ø³ØªÙŠØ±Ø§Ø¯ Ø§Ù„Ù…ÙƒØªØ¨Ø§Øª
# ======================
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import normalize
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Conv2DTranspose, concatenate
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import load_img, img_to_array

# ======================
# 2. Ù…Ø³Ø§Ø±Ø§Øª Ø§Ù„ØµÙˆØ± ÙˆØ§Ù„Ù…Ø§Ø³ÙƒØ§Øª
# ======================
images_path = r"C:\Users\ASAA\Downloads\project_extract\data\images"
masks_path  = r"C:\Users\ASAA\Downloads\project_extract\data\masks"

IMG_HEIGHT = 128
IMG_WIDTH  = 128
IMG_CHANNELS = 3

# ======================
# 3. ØªØ­Ù…ÙŠÙ„ Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª
# ======================
images = []
masks = []

for file in os.listdir(images_path):
    img = load_img(os.path.join(images_path, file), target_size=(IMG_HEIGHT, IMG_WIDTH))
    img = img_to_array(img)
    images.append(img)

    # Ø§Ù„Ù…Ø§Ø³Ùƒ Ù„ÙŠÙ‡ Ù†ÙØ³ Ø§Ù„Ø§Ø³Ù… + ÙƒÙ„Ù…Ø© mask
    if file.endswith(".tif"):
        mask_file = file.replace(".tif", "_mask.tif")
    elif file.endswith(".png"):
        mask_file = file.replace(".png", "_mask.png")
    else:
        continue  # Ù„Ùˆ ØµÙŠØºØ© ØºÙŠØ± Ù…Ø¯Ø¹ÙˆÙ…Ø©ØŒ Ù†ØªØ¬Ø§Ù‡Ù„Ù‡Ø§

    mask_path = os.path.join(masks_path, mask_file)
    if os.path.exists(mask_path):
        mask = load_img(mask_path, target_size=(IMG_HEIGHT, IMG_WIDTH), color_mode="grayscale")
        mask = img_to_array(mask)
        masks.append(mask)

images = np.array(images) / 255.0
masks = np.array(masks) / 255.0
masks = (masks > 0.5).astype(np.uint8)   # ØªØ­ÙˆÙŠÙ„ Ø§Ù„Ù…Ø§Ø³Ùƒ Ù„Ù€ 0 Ùˆ 1 ÙÙ‚Ø·

print("âœ… Ø­Ø¬Ù… Ø§Ù„ØµÙˆØ±:", images.shape)
print("âœ… Ø­Ø¬Ù… Ø§Ù„Ù…Ø§Ø³ÙƒØ§Øª:", masks.shape)

# ======================
# 4. ØªÙ‚Ø³ÙŠÙ… Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª (Train/Test)
# ======================
X_train, X_test, y_train, y_test = train_test_split(images, masks, test_size=0.2, random_state=42)

print("ğŸ”¹ ØªØ¯Ø±ÙŠØ¨:", X_train.shape, y_train.shape)
print("ğŸ”¹ Ø§Ø®ØªØ¨Ø§Ø±:", X_test.shape, y_test.shape)

# ======================
# 5. Ø¨Ù†Ø§Ø¡ U-Net
# ======================
def unet_model(input_size=(IMG_HEIGHT, IMG_WIDTH, IMG_CHANNELS)):
    inputs = Input(input_size)

    # Encoder
    c1 = Conv2D(16, (3,3), activation="relu", padding="same")(inputs)
    c1 = Conv2D(16, (3,3), activation="relu", padding="same")(c1)
    p1 = MaxPooling2D((2,2))(c1)

    c2 = Conv2D(32, (3,3), activation="relu", padding="same")(p1)
    c2 = Conv2D(32, (3,3), activation="relu", padding="same")(c2)
    p2 = MaxPooling2D((2,2))(c2)

    c3 = Conv2D(64, (3,3), activation="relu", padding="same")(p2)
    c3 = Conv2D(64, (3,3), activation="relu", padding="same")(c3)
    p3 = MaxPooling2D((2,2))(c3)

    c4 = Conv2D(128, (3,3), activation="relu", padding="same")(p3)
    c4 = Conv2D(128, (3,3), activation="relu", padding="same")(c4)
    p4 = MaxPooling2D((2,2))(c4)

    # Bottleneck
    c5 = Conv2D(256, (3,3), activation="relu", padding="same")(p4)
    c5 = Conv2D(256, (3,3), activation="relu", padding="same")(c5)

    # Decoder
    u6 = Conv2DTranspose(128, (2,2), strides=(2,2), padding="same")(c5)
    u6 = concatenate([u6, c4])
    c6 = Conv2D(128, (3,3), activation="relu", padding="same")(u6)
    c6 = Conv2D(128, (3,3), activation="relu", padding="same")(c6)

    u7 = Conv2DTranspose(64, (2,2), strides=(2,2), padding="same")(c6)
    u7 = concatenate([u7, c3])
    c7 = Conv2D(64, (3,3), activation="relu", padding="same")(u7)
    c7 = Conv2D(64, (3,3), activation="relu", padding="same")(c7)

    u8 = Conv2DTranspose(32, (2,2), strides=(2,2), padding="same")(c7)
    u8 = concatenate([u8, c2])
    c8 = Conv2D(32, (3,3), activation="relu", padding="same")(u8)
    c8 = Conv2D(32, (3,3), activation="relu", padding="same")(c8)

    u9 = Conv2DTranspose(16, (2,2), strides=(2,2), padding="same")(c8)
    u9 = concatenate([u9, c1])
    c9 = Conv2D(16, (3,3), activation="relu", padding="same")(u9)
    c9 = Conv2D(16, (3,3), activation="relu", padding="same")(c9)

    outputs = Conv2D(1, (1,1), activation="sigmoid")(c9)

    model = Model(inputs, outputs)
    return model

model = unet_model()
model.compile(optimizer=Adam(), loss="binary_crossentropy", metrics=["accuracy"])
model.summary()

# ======================
# 6. Ø§Ù„ØªØ¯Ø±ÙŠØ¨ (Ø¹Ø¯Ø¯ Epochs = 5)
# ======================
history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=5,        # ğŸ‘ˆ Ø§ØªØºÙŠØ±Øª Ù…Ù† 10 Ø¥Ù„Ù‰ 5
    batch_size=16
)

# ======================
# 7. Ø­ÙØ¸ Ø§Ù„Ù…ÙˆØ¯ÙŠÙ„
# ======================
model.save("unet_segmentation.h5")
print("ğŸ’¾ ØªÙ… Ø­ÙØ¸ Ø§Ù„Ù…ÙˆØ¯ÙŠÙ„ ÙÙŠ: unet_segmentation.h5")

# ======================
# 8. Ø§Ø®ØªØ¨Ø§Ø± Ø§Ù„ØªÙ†Ø¨Ø¤
# ======================
idx = 5
test_img = X_test[idx]
true_mask = y_test[idx]
pred_mask = model.predict(np.expand_dims(test_img, axis=0))[0]

plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.title("Input Image")
plt.imshow(test_img)

plt.subplot(1,3,2)
plt.title("True Mask")
plt.imshow(true_mask[:,:,0], cmap="gray")

plt.subplot(1,3,3)
plt.title("Predicted Mask")
plt.imshow(pred_mask[:,:,0] > 0.5, cmap="gray")

plt.show()
```

```python
import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam

# Ø§Ù„Ù…Ø³Ø§Ø±Ø§Øª
train_dir = r"C:\Users\ASAA\Downloads\project_extract\data\images"
mask_dir = r"C:\Users\ASAA\Downloads\project_extract\data\masks"
models_dir = r"C:\Users\ASAA\Downloads\BrainTumorApp\models"

os.makedirs(models_dir, exist_ok=True)

# Ø¥Ø¹Ø¯Ø§Ø¯ Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª (Ù†Ø¹Ù… / Ù„Ø§)
datagen = ImageDataGenerator(validation_split=0.2, rescale=1./255)

train_gen = datagen.flow_from_directory(
    train_dir,
    target_size=(224,224),
    batch_size=32,
    class_mode='binary',
    subset='training'
)

val_gen = datagen.flow_from_directory(
    train_dir,
    target_size=(224,224),
    batch_size=32,
    class_mode='binary',
    subset='validation'
)

# Ø¨Ù†Ø§Ø¡ Ø§Ù„Ù…ÙˆØ¯ÙŠÙ„
base_model = EfficientNetB0(include_top=False, weights="imagenet", input_shape=(224,224,3))
x = GlobalAveragePooling2D()(base_model.output)
x = Dropout(0.5)(x)
out = Dense(1, activation="sigmoid")(x)
classifier = Model(base_model.input, out)

classifier.compile(optimizer=Adam(1e-4), loss="binary_crossentropy", metrics=["accuracy"])

# ØªØ¯Ø±ÙŠØ¨
classifier.fit(train_gen, validation_data=val_gen, epochs=5)

# Ø­ÙØ¸ Ø§Ù„Ù…ÙˆØ¯ÙŠÙ„
classifier.save(os.path.join(models_dir, "classifier.h5"))
print("âœ… ØªÙ… Ø­ÙØ¸ Ø§Ù„Ù…ÙˆØ¯ÙŠÙ„: classifier.h5")
```
